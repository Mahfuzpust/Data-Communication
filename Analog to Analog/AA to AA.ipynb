# AM, FM,PM

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Amplitude modulation carrier_signal
ac = 2
fc = 5
pc = 0

time = np.arange(0,10, .001)

carrier_signal = ac*np.sin(2*np.pi*fc*time + pc)

plt.plot(time, carrier_signal)
plt.show()
```

```python
#Amplitude modulation message_signal

am = .2
fm = .8
pm = 0

message_signal = am*np.sin(2*np.pi*fm*time + pm)

plt.plot(time, message_signal)
plt.show()
```

```python
#Amplitude modulation 

signal = carrier_signal * message_signal

plt.plot(time, signal)
plt.show()
```

```python
#Frequency Modulation carrier_signal

import matplotlib.pyplot as plt
import numpy as np

ac = 2
fc = 5
pc = 0

time = np.arange(0,10, .001)

carrier_signal = ac*np.sin(2*np.pi*fc*time + pc)

plt.plot(time, carrier_signal)
plt.show()

```

```python
#Frequency Modulation message_signal

am = .2
fm = .8
pm = 0

message_signal = am*np.sin(2*np.pi*fm*time + pm)

plt.plot(time, message_signal)
plt.show()
```

```python
#Frequency Modulation

frequency_modulation = ac*np.sin(2*np.pi*(fc+message_signal)*time +pc)

plt.plot(time, frequency_modulation)
plt.show()
```

```python
#Phase Modulation carrier_signal

import matplotlib.pyplot as plt
import numpy as np

ac = 2
fc = 5
pc = 0

time = np.arange(0,10, .1)

carrier_signal = ac*np.sin(2*np.pi*fc*time + pc)

plt.plot(time, carrier_signal)
plt.show()
```

```python
am = .2
fm = .8
pm = 0

message_signal = am*np.sin(2*np.pi*fm*time + pm)

plt.plot(time, message_signal)
plt.show()
```

```python
phase_modulation = np.sin(2*np.pi*fc*time + message_signal)
plt.plot(time, phase_modulation)
plt.show()
```
